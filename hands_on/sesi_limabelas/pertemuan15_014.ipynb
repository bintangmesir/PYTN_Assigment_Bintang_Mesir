```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
from sklearn.compose import ColumnTransformer
```

```python
tips = pd.get_dummies(df[['sex', 'smoker', 'time']])
tips.head()
```

```python
tips = pd.concat([df_enc, tips], axis = 1)
tips.head()
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
robust = RobustScaler()
```

```python
tips['total_bill'] = robust.fit_transform(df[['total_bill']])
```

```python
df.head()
```

```python
tips['tip'] = df['tip']
tips.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = tips.drop(columns = 'tip')
y = tips['tip']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size = 0.3)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
from sklearn.impute import SimpleImputer
```

```python
df = pd.DataFrame({
    'x1':[4,5,np.nan,6,7,9],
    'x2':[3,5,6,5,np.nan,5],
    'x3':[10,11,12,9,8,11],
    'x4':['A','A','C','C','D',np.nan],
    'x5':['X','Y','X','X',np.nan,'Y'],
    'x6':['M','M',np.nan,'M','N',np.nan]
    })
```

```python
df
```

```python
imp = SimpleImputer(strategy = 'median')
df[['x1', 'x2', 'x3']] = imp.fit_transform(df[['x1', 'x2', 'x3']])
df.head()
```

```python
imp_mode = SimpleImputer(strategy = 'most_frequent')
df[['x4', 'x5']] = imp_mode.fit_transform(df[['x4', 'x5']])
df.head()
```

```python
imp_constant = SimpleImputer(strategy = 'constant', fill_value = 'N')
df[['x6']] = imp_constant.fit_transform(df[['x6']])
df.head()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```
